```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
#Params:
r = 27
sig = 22.12
beta = 0.19
```

### The equations

$\frac{dx}{dt}= \sigma(y-x),$ \
$\frac{dy}{dt}= x(r-z)-y,$ \
$\frac{dz}{dt}= xy- \beta z ,$

The Jacobian can be easily derived:
$
J =
\begin{pmatrix}
-\sigma & \sigma & 0  \\
r-z & -1 & -x  \\
y  & x  & -\beta   
\end{pmatrix}
$

```python
def f(t,q):
    return np.array([
        sig*(q[1]-q[0]),
        q[0]*(r-q[2])-q[1],
        q[0]*q[1]-beta*q[2]
    ])
```

```python
#inital conditions:
q0 = [int(1),int(1),int(1)]
tspan = (0,4000)
sol1 = solve_ivp(f, tspan, q0, method='RK45', t_eval=np.linspace(0, 100, 40000))
```
